```python
#Import
import pandas as pd
import time
import logging
import sys
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from datetime import datetime
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--incognito")
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)
```

```python
#Define specialty and city resource files
dfSpecialties= pd.read_csv(r'.\specialties-01.csv')
#dfCities= pd.read_csv(r'.\cities.csv')
dfCities= pd.read_csv(r'.\cities-01.csv')
```

```python
#Define count constant
logger = logging.getLogger()
count = 1000
countClicScrollBy = 15
shortTimeSleep = 3
longTimeSleep = 30
```

```python
#Method to record the logs
def log(e):
    %run -i notification-caution_telegram.ipynb
    logger.exception('Exception occurred while code execution en script Agenda Medica Ecuador: ' + str(e))

#Method search
try:
    for i in range(count):
        print ('')
        print (i+1, '- periodo |', datetime.today().strftime('%Y-%m-%d %H:%M:%S'))
        for row, specialty in dfSpecialties.iterrows():
            nameSpecialty = specialty['nombre'].lower()
            for row, city in dfCities.iterrows():
                nameCity = city['nombre'].lower()                           
                
                # Enter the google page
                driver.get('https://www.google.com/')
                time.sleep(shortTimeSleep)
                # driver.maximize_window()
                
                # Enter specialty and city data in google
                try:
                    driver.find_element(By.XPATH,"//textarea[contains(@title,'Buscar')]").send_keys('agenda medica ecuador ', nameSpecialty, ' ', nameCity)
                    driver.find_element(By.XPATH,"//textarea[contains(@title,'Buscar')]").submit()
                except Exception as e:
                    log(e)
                    continue
                    
                print(nameSpecialty, '-', nameCity)
                ############################################################# 
                #                 SELECT RESULT ON GOOGLE                   #
                ############################################################# 
                try:
                    driver.find_element(By.XPATH, "//a[contains(@href,'https://www.agendamedicaecuador.com/medicos')]").click()                   
                except Exception as e:
                    log(e)
                    continue
                    
                time.sleep(shortTimeSleep)
                
                ############################################################# 
                #            ENTER CITY AND SPECIALTY EN AME                #
                ############################################################# 

                # Enter the parameter of the specialty to search
                driver.find_element(By.XPATH,"//input[contains(@placeholder,'Ej: Especialidad')]").send_keys(nameSpecialty)
                time.sleep(shortTimeSleep)

                # Select specialty
                driver.find_element("xpath",'/html/body/div/div/div/main/div/div/div/form/div[1]/div[1]/div[1]/div[2]/div/ul/li[1]').click()
                time.sleep(shortTimeSleep)

                # Enter the parameter of the city to search
                driver.find_element(By.XPATH,"//input[contains(@placeholder,'Ej: Quito')]").send_keys(nameCity)
                time.sleep(shortTimeSleep)

                # Select city
                driver.find_element("xpath",'/html/body/div/div/div/main/div/div/div/form/div[1]/div[1]/div[2]/div[2]/div/ul/li[1]').click()
                time.sleep(shortTimeSleep)

                ############################################################# 
                #                         SCROLL MOUSE                      #
                ############################################################# 
                for c in range(countClicScrollBy):
                    time.sleep(1)
                    driver.execute_script("window.scrollBy(0, 100);")
                time.sleep(longTimeSleep)
                
except Exception as e:
    %run -i notification-error_telegram.ipynb
    logger.exception('Exception occurred while code execution en script Agenda Medica Ecuador: ' + str(e))
            
    
```
